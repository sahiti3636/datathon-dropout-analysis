```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
enr1 = pd.read_csv("100_enr1.csv")
enr2 = pd.read_csv("100_enr12.csv")
prof1 = pd.read_csv("100_prof1.csv")
prof2 = pd.read_csv("100_prof12.csv")
```

```python
prof1["rural_urban"] = prof1["rural_urban"].map({1: "Rural", 2: "Urban"})
prof2["rural_urban"] = prof2["rural_urban"].map({1: "Rural", 2: "Urban"})
```

```python
mapping = {
    (1, 1): "General",
    (1, 2): "SC",
    (1, 3): "ST",
    (1, 4): "OBC",
    (2, 5): "Muslim",
    (2, 6): "Christian",
    (2, 7): "Sikh",
    (2, 8): "Buddhist",
    (2, 9): "Parsi",
    (2, 10): "Jain"
}
```

```python
enr1["caste"] = enr1.apply(
    lambda row: mapping.get((row["item_group"], row["item_id"])), axis=1
)
```

```python
enr2["caste"] = enr2.apply(
    lambda row: mapping.get((row["item_group"], row["item_id"])), axis=1
)
```

```python
enr1 = enr1.dropna(subset=["caste"]).reset_index(drop=True)
enr2 = enr2.dropna(subset=["caste"]).reset_index(drop=True)
```

```python
enr1 = enr1.drop(columns=["pseudocode", "item_group", "item_id"], errors="ignore")
enr2 = enr2.drop(columns=["pseudocode", "item_group", "item_id"], errors="ignore")
```

```python
numeric_cols = [col for col in enr1.columns if col not in ['caste'] and enr1[col].dtype in ['int64', 'float64']]
```

```python
aggregated_by_caste_1 = enr1.groupby('caste')[numeric_cols].sum().reset_index()
```

```python
aggregated_by_caste_2 = enr2.groupby('caste')[numeric_cols].sum().reset_index()
```

```python
aggregated_by_caste_1
```

```python
aggregated_by_caste_1.columns
```

```python
aggregated_by_caste_2
```

```python
merged = pd.merge(
    aggregated_by_caste_1,
    aggregated_by_caste_2,
    on="caste",
    suffixes=("_23_24", "_24_25")
)
```

```python
result = merged[["caste"]].copy()
sequence = ["cpp"] + [f"c{i}" for i in range(1, 13)]
```

```python
for i in range(1, len(sequence)):
    prev_grade = sequence[i - 1]
    curr_grade = sequence[i]

    for gender in ["b", "g"]:
        prev_col = f"{prev_grade}_{gender}_23_24"
        curr_col = f"{curr_grade}_{gender}_24_25"

        dropout_col = f"{curr_grade}_{gender}_dropout_rate"
        retention_col = f"{curr_grade}_{gender}_retention_rate"

        # Dropout rate formula
        result[dropout_col] = (
            (merged[prev_col] - merged[curr_col]) / merged[prev_col]
        ).fillna(0)

        # Clip negative values (no negative dropout)
        result[dropout_col] = result[dropout_col].clip(lower=0)

        # Retention = 1 - dropout
        result[retention_col] = 1 - result[dropout_col]
```

```python
result.to_csv("caste_dropout_retention_rates.csv", index=False)
```
